
# Awareness Is Key: Other Dictionary Functions

## Accessing Values

### Definition  
Access a value in a dictionary by using square brackets with the key or use `.get()` for safer access with an optional default value.

### Example Code
```python
financial_data = {"revenue": 100000, "expenses": 75000}

# Access value using square brackets
revenue = financial_data["revenue"]
print(revenue)

# Access value using .get() with default value
taxes = financial_data.get("taxes", 0)
print(taxes)
```

### Explanation  
- `financial_data["revenue"]` returns the value linked to `"revenue"`.  
- `.get("taxes", 0)` returns `0` because `"taxes"` is not in the dictionary.  
- `.get()` avoids an error if the key does not exist.


---

## Removing Entries

### Definition  
Remove entries from a dictionary using `.pop()`, `.popitem()`, `del`, or `.clear()`.

### Example Code
```python
financial_data = {"revenue": 100000, "expenses": 75000, "assets": 500000}

# Remove specific item by key
financial_data.pop("expenses")

# Remove the most recent item
financial_data.popitem()

# Delete by key using del
del financial_data["revenue"]

# Remove all entries
financial_data.clear()

print(financial_data)
```

### Explanation  
- `.pop("expenses")` removes the entry with key `"expenses"`.  
- `.popitem()` removes the last inserted key–value pair.  
- `del financial_data["revenue"]` deletes the entry with key `"revenue"`.  
- `.clear()` removes all entries, leaving an empty dictionary.

---

## Retrieving Dictionary Contents

### Definition  
Get the key–value pairs from a dictionary using `.keys()` and `.values()`.

### Example Code
```python
financial_data = {"revenue": 100000, "expenses": 75000}

print(financial_data.items())   # dict_items([('revenue', 100000), ('expenses', 75000)])
```

### Explanation  
- `.items()` returns all key–value pairs as tuples.


---

## Merging Dictionaries

### Definition  
Use `.update()` to merge one dictionary into another. Existing keys are updated, and new keys are added.

### Example Code
```python
financial_data = {"revenue": 100000, "expenses": 75000}
new_data = {"assets": 500000, "expenses": 80000}

financial_data.update(new_data)
print(financial_data)
```

### Explanation  
- `"expenses"` is updated to `80000`.  
- `"assets"` is added as a new key.
