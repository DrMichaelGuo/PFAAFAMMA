# Follow-Along Activity

**Note**: When running this example, all steps must be executed together in a single code cell. If the code is split into separate cells, the 3D plot may not display because Matplotlib does not retain the figure between steps. Always run the complete code block from start to finish to generate the correct output.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Create grid data for pricing (X) and sales volume (Y)
X = np.linspace(0, 10, 50)
Y = np.linspace(0, 10, 50)
X, Y = np.meshgrid(X, Y)
# Sample function for profit: a sinusoidal function for demonstration
Z = np.sin(np.sqrt(X**2 + Y**2)) * 10 + 50

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title("Surface Plot: Profit Surface")
ax.set_xlabel("Pricing")
ax.set_ylabel("Sales Volume")
ax.set_zlabel("Profit")
fig.colorbar(surf, shrink=0.5, aspect=5)

# Save to jpeg for textbook publication
plt.savefig('surface_plot.jpeg', dpi=300, bbox_inches='tight')

plt.show()

```
