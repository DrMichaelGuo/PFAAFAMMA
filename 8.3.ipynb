# Follow-Along Activity

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
```

```python
# Create grid data for pricing (X) and sales volume (Y)
X = np.linspace(0, 10, 50)
Y = np.linspace(0, 10, 50)
X, Y = np.meshgrid(X, Y)
# Sample function for profit: a sinusoidal function for demonstration
Z = np.sin(np.sqrt(X**2 + Y**2)) * 10 + 50
```

```python
# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title("Surface Plot: Profit Surface")
ax.set_xlabel("Pricing")
ax.set_ylabel("Sales Volume")
ax.set_zlabel("Profit")
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
