# Follow-Along Activity

```python
import pandas as pd
from openpyxl import load_workbook
from google.colab import files
```

```python
uploaded = files.upload()
df = pd.read_excel('9_excel_example.xlsx', sheet_name='Sheet1')
print(df.head())
```

```python
summary = df.groupby('Customer')['Amount'].sum().reset_index()
summary.columns = ['Customer', 'Total Amount']

with pd.ExcelWriter('9_excel_summary.xlsx', engine='openpyxl', mode='w') as writer:
    df.to_excel(writer, sheet_name='Sheet1', index=False)
    summary.to_excel(writer, sheet_name='Summary', index=False)

files.download('9_excel_summary.xlsx')
```

```python
wb = load_workbook('9_excel_summary.xlsx')
ws = wb['Sheet1']
ws['B1'] = 'Processed'
wb.save('9_excel_updated.xlsx')
files.download('9_excel_updated.xlsx')
```

# Your Project

```python
# Import libraries
import pandas as pd
from openpyxl import load_workbook
from google.colab import files

# Upload and read the Excel file
uploaded = files.upload()
df = pd.read_excel('9_excel_project.xlsx', sheet_name='Data')
print(df.head())

# Calculate total sales per customer
summary = df.groupby('Customer')['Sales'].sum().reset_index()
summary.columns = ['Customer', 'Total Sales']

# Write both raw data and summary to a new Excel file
with pd.ExcelWriter('9_excel_project_output.xlsx', engine='openpyxl', mode='w') as writer:
    df.to_excel(writer, sheet_name='Data', index=False)
    summary.to_excel(writer, sheet_name='Summary', index=False)

files.download('9_excel_project_output.xlsx')

# Update cell A1 in the Summary sheet
wb = load_workbook('9_excel_project_output.xlsx')
ws = wb['Summary']
ws['A1'] = 'Sales Summary'
wb.save('9_excel_project_final.xlsx')

# Download the updated file
files.download('9_excel_project_final.xlsx')
```
