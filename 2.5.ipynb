## Your Project

```python
# Step 1: Capture user input for total sales and total costs
# input() prompts the user and stores the values as strings
total_sales = input("Enter total sales (£): ")
total_costs = input("Enter total costs (£): ")

# Step 2: Convert inputs to floats
# Convert the string inputs to float for numeric calculation
total_sales = float(total_sales)
total_costs = float(total_costs)

# Step 3: Calculate gross profit
# Gross profit is the difference between total sales and total costs
gross_profit = total_sales - total_costs

# Step 4: Calculate profit margin
# Profit margin is calculated as a percentage of total sales
# Formula: (Gross Profit / Total Sales) * 100
profit_margin = (gross_profit / total_sales) * 100

# Step 5: Display the results with formatting
# Use f-strings to display the results rounded to two decimal places
print(f"Gross Profit (£): {gross_profit:.2f}")
print(f"Profit Margin (%): {profit_margin:.2f}")
```
