# Follow-Along Activity

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate sample data for 3 departments
np.random.seed(42)
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 15, 100)
data3 = np.random.normal(55, 20, 100)
data = [data1, data2, data3]
```

```python
# Create the violin plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.violinplot(data, showmeans=True)
ax.set_title("Violin Plot of Monthly Revenue for 3 Departments")
ax.set_xlabel("Department")
ax.set_ylabel("Monthly Revenue (Â£'000s)")
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(["Dept 1", "Dept 2", "Dept 3"])

# Save to jpeg for textbook publication
plt.savefig("violin_plot.jpeg", dpi=300)
plt.show()
```
