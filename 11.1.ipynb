```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define the ticker symbol for Apple Inc.
ticker_symbol = 'AAPL'

# Create a Ticker object using yfinance
ticker = yf.Ticker(ticker_symbol)

# Retrieve one year of historical data
data = ticker.history(period='1y')

print(data.head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='AAPL Closing Price')
plt.title('Apple Inc. Closing Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)

# Save to jpeg for tetxtbook publication
plt.savefig('Apple Inc. Closing Price.jpg', format='jpg', dpi=300)

plt.show()
```

```python
# Calculate daily percentage change
data['Daily Return'] = data['Close'].pct_change()

print(data[['Daily Return']].head())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Daily Return'], label='AAPL Daily Return', color='purple')
plt.title('Apple Inc. Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.grid(True)

# Save to jpeg for tetxtbook publication
plt.savefig('Apple Inc. Daily Returns.jpg', format='jpg', dpi=300)

plt.show()
```

```python
# Install and upgrade yfinance to the latest version to avoid errors when downloading data
!pip install --upgrade yfinance

# Import necessary libraries
import yfinance as yf  # For accessing financial data
import pandas as pd     # For working with data in tabular form
import matplotlib.pyplot as plt  # For creating plots
```

```python
# Retrieve Historical Stock Data
ticker_symbol = 'GOOGL'
time_period = '1y'  # Choose time range ('6mo', '1y', '2y', etc.)

# Download stock data
ticker = yf.Ticker(ticker_symbol)
data = ticker.history(period=time_period)

# Display the first few rows of the dataset
print(data.head())
```

```python
# Data Analysis and Visualisation (Closing Price)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label=f'{ticker_symbol} Closing Price', color='blue')
plt.title(f'{ticker_symbol} Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate daily return percentage
data['Daily Return'] = data['Close'].pct_change()

print(data[['Daily Return']].head())
```

```python
# Plot Daily Returns
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Daily Return'], label=f'{ticker_symbol} Daily Return', color='purple')
plt.title(f'{ticker_symbol} Daily Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.grid(True)
plt.show()
```

## Extra: GBP to USD exchange rate over 3 months
