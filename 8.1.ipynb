## Follow-Along Activity

### Step1: Line Chart

```python
# Importing matplotlib and other useful libraries
import matplotlib.pyplot as plt # alias

# Sample data for monthly revenue (in £)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [10000, 12000, 9000, 15000, 13000, 17000]

# Creating a figure and plotting the line chart
plt.figure(figsize=(8, 4))
plt.plot(months, revenue, marker='o', linestyle='-', color='blue', label='Revenue')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.title('Monthly Revenue')

# Adding a grid and legend for clarity
plt.grid(True)
plt.legend()

# Save the plot as JPEG for textbook publication
plt.savefig('monthly_revenue.jpg', dpi=300, format='jpeg')

# Display the plot
plt.show()

```

### Step 2: Bar Chart

```python
# Sample data for monthly expenses (in £)
categories = ['Rent', 'Utilities', 'Salaries', 'Marketing', 'Miscellaneous']
expenses = [3000, 1200, 7000, 1500, 800]

# Creating the bar chart
plt.figure(figsize=(8, 4))
plt.bar(categories, expenses, color='green')

# Adding labels and title
plt.xlabel('Expense Category')
plt.ylabel('Amount (£)')
plt.title('Monthly Expenses')

# Save the plot as JPEG for textbook publication
plt.savefig('monthly_expenses.jpg', dpi=300, format='jpeg')

# Display the plot
plt.show()


```

### Step 3: Pie Chart

```python
# Data: Expense categories and their corresponding amounts (£)
categories = ['Rent', 'Utilities', 'Salaries', 'Marketing', 'Miscellaneous']
expenses = [3000, 1200, 7000, 1500, 800]

# Create the pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    expenses,
    labels=categories,
    autopct='%1.1f%%',
    startangle=140,
    colors=['lightblue', 'orange', 'lightgreen', 'red', 'purple'],
    explode=[0, 0, 0.1, 0, 0]  # Slightly separates the 'Salaries' slice
)

# Add a title
plt.title('Monthly Expenses Breakdown')

# Save the plot as JPEG for textbook publication
plt.savefig('monthly_expenses_breakdown.jpg', dpi=300, format='jpeg')

# Display the pie chart
plt.show()
```

## Task 1: Modifying the Line Chart

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [10000, 12000, 9000, 15000, 13000, 17000]

# Resize the figure to 10 inches wide and 6 inches high
plt.figure(figsize=(10, 6))

"""
Restyle the line chart by updating the plot() function with the following arguments:
marker='s'          # square markers
linestyle='--'      # dashed line
colour='green'      # change colour
"""
plt.plot(months, revenue, marker='s', linestyle='--', color='green', label='Revenue')

plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.title('Monthly Revenue')
plt.grid(True)
plt.legend()
plt.show()
```

## Task 2: Modifying the Bar Chart (Horizontal)

```python
categories = ['Rent', 'Utilities', 'Salaries', 'Marketing', 'Miscellaneous']
expenses = [3000, 1200, 7000, 1500, 800]

plt.figure(figsize=(8, 4))

# Switch to a horizontal layout by replacing plt.bar() with plt.barh()
# Change the bar colour to orange by setting: colour='orange'
plt.barh(categories, expenses, color='orange')

plt.xlabel('Amount (£)')
plt.ylabel('Expense Category')
plt.title('Monthly Expenses (Horizontal Bar Chart)')

plt.show()
```

## Task 3: Modifying the Pie Chart

```python
categories = ['Rent', 'Utilities', 'Salaries', 'Marketing', 'Miscellaneous']
expenses = [3000, 1200, 7000, 1500, 800]

plt.figure(figsize=(7, 7))

plt.pie(expenses,
        labels=categories,
        autopct='%1.1f%%',
        startangle=90, # Rotate the chart by setting startangle=90 in the plt.pie() function.
        colors=['lightblue', 'orange', 'lightgreen', 'red', 'purple'],
        explode=[0.1, 0, 0, 0, 0]) # 10 % gap on the first slice

plt.title('Monthly Expenses Breakdown')

plt.show()
```
