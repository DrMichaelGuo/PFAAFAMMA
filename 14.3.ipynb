# Follow-Along Activity A

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Step 2
initial_investment = 1_000
years = 10
mean_return = 0.05
std_deviation = 0.02
simulations = 1_000
```

```python
final_values = []

for simulation in range(simulations):
    value = initial_investment
    for year in range(years):
        annual_return = np.random.normal(mean_return, std_deviation)
        value = value * (1 + annual_return)
    final_values.append(value)
```

```python
plt.hist(final_values, bins=50)
plt.title("Simulated investment value after 10 years")
plt.xlabel("Final value (£)")
plt.ylabel("Frequency")
plt.grid(True)

# Save to jpeg for textbook publication purposes
plt.savefig("Figure 14.11 Simulated investment value after 10 years.jpeg", dpi=300)

plt.show()
```

```python
mean_value = np.mean(final_values)
percentile_5 = np.percentile(final_values, 5)
percentile_95 = np.percentile(final_values, 95)
loss_probability = np.mean(np.array(final_values) < initial_investment)

print(f"Mean final value: £{mean_value:.2f}")
print(f"5th percentile  : £{percentile_5:.2f}")
print(f"95th percentile : £{percentile_95:.2f}")
print(f"Probability of losing money: {loss_probability:.2f}")

```

# Follow-Along Activity B

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
# Step 2
S0 = 100
K = 105
r = 0.05
sigma = 0.2
T = 1
simulations = 10000
```

```python
Z = np.random.normal(0, 1, simulations)
ST = S0 * np.exp((r - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * Z)
```

```python
payoffs = np.maximum(ST - K, 0)
```

```python
option_price = np.exp(-r * T) * np.mean(payoffs)
print(f"Estimated European Call Option Price: £{option_price:.2f}")
```

```python
plt.hist(ST, bins=50, edgecolor='black')
plt.title("Simulated Stock Prices at Expiry")
plt.xlabel("Stock Price (£)")
plt.ylabel("Frequency")
plt.grid(True)

# Save to jpeg for textbook publication purposes
plt.savefig("Figure 15.13 Simulated Stock Prices at Expiry.jpeg", dpi=300, bbox_inches='tight')

plt.show()
```

```python
print("----- Summary Statistics -----")
print(f"Mean final stock price: £{np.mean(ST):.2f}")
print(f"Standard deviation of stock prices: £{np.std(ST):.2f}")
print(f"Mean payoff: £{np.mean(payoffs):.2f}")
print(f"Probability of positive payoff: {np.mean(payoffs > 0):.2f}")
print(f"Maximum payoff: £{np.max(payoffs):.2f}")
print(f"Minimum payoff: £{np.min(payoffs):.2f}")
```

```python
plt.hist(payoffs, bins=50, edgecolor='black')
plt.title("Distribution of Option Payoffs")
plt.xlabel("Payoff (£)")
plt.ylabel("Frequency")
plt.grid(True)

# Save to jpeg for textbook publication purposes
plt.savefig("Figure 14.14 Distribution of Option Payoffs.jpeg", dpi=300, bbox_inches='tight')

plt.show()
```
