```python
# Upload CSV file
from google.colab import files
uploaded = files.upload()

# Read the data
import pandas as pd
df = pd.read_csv('6_data.csv')
```

```python
# Select 'Revenue' column
revenue = df['Revenue']

# Central tendency
print("Mean:", round(revenue.mean(), 2))
print("Median:", round(revenue.median(), 2))
print("Mode:", round(revenue.mode().iloc[0], 2))
```

```python
# Dispersion
print("Standard Deviation:", round(revenue.std(), 2))
print("Variance:", round(revenue.var(), 2))
print("Minimum:", round(revenue.min(), 2))
print("Maximum:", round(revenue.max(), 2))
print("1st Quartile (Q1):", round(revenue.quantile(0.25), 2))
print("3rd Quartile (Q3):", round(revenue.quantile(0.75), 2))
print("Range (Max - Min):", round(revenue.max() - revenue.min(), 2))
print("Interquartile Range (Q3 - Q1):", round(revenue.quantile(0.75) - revenue.quantile(0.25), 2))
```

```python
# Shape of distribution
print("Skewness:", round(revenue.skew(), 2))
print("Kurtosis:", round(revenue.kurt(), 2))
```

```python
# Upload CSV file
from google.colab import files
uploaded = files.upload()

# Read the data
import pandas as pd
df = pd.read_csv('6_data.csv')

# Select 'Expenses' column
expenses = df['Expenses']

# Central tendency
print("Mean:", round(expenses.mean(), 2))
print("Median:", round(expenses.median(), 2))
print("Mode:", round(expenses.mode().iloc[0], 2))

# Dispersion
print("Standard Deviation:", round(expenses.std(), 2))
print("Variance:", round(expenses.var(), 2))
print("Minimum:", round(expenses.min(), 2))
print("Maximum:", round(expenses.max(), 2))
print("1st Quartile:", round(expenses.quantile(0.25), 2))
print("3rd Quartile:", round(expenses.quantile(0.75), 2))
print("Range:", round(expenses.max() - expenses.min(), 2))
print("Interquartile Range:", round(expenses.quantile(0.75) - expenses.quantile(0.25), 2))

# Shape of distribution
print("Skewness:", round(expenses.skew(), 2))
print("Kurtosis:", round(expenses.kurt(), 2))
```
