```python
import pandas as pd

# Dataset with monthly revenue data for three years (2021-2023)
data = {
    'Date': [
        '2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01',
        '2021-05-01', '2021-06-01', '2021-07-01', '2021-08-01',
        '2021-09-01', '2021-10-01', '2021-11-01', '2021-12-01',
        '2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01',
        '2022-05-01', '2022-06-01', '2022-07-01', '2022-08-01',
        '2022-09-01', '2022-10-01', '2022-11-01', '2022-12-01',
        '2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01',
        '2023-05-01', '2023-06-01', '2023-07-01', '2023-08-01',
        '2023-09-01', '2023-10-01', '2023-11-01', '2023-12-01'
    ],
    'Revenue': [
        250000, 260000, 270000, 280000,
        290000, 300000, 310000, 320000,
        330000, 340000, 350000, 360000,
        270000, 280000, 290000, 300000,
        310000, 320000, 330000, 340000,
        350000, 360000, 370000, 380000,
        300000, 320000, 290000, 310000,
        305000, 330000, 340000, 360000,
        350000, 370000, 380000, 390000
    ]
}

# Load dataset to dataframe
df = pd.DataFrame(data)
```

```python
# Converting Date column to datetime format and setting it as index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
print(df.head())
```

```python
# Plotting monthly revenue over time
df['Revenue'].plot(title="Monthly Revenue Over Time", grid=True)
```

```python
# Resampling data to get quarterly revenue
quarterly_revenue = df['Revenue'].resample('QE').sum()
print(quarterly_revenue)
```

```python
quarterly_revenue.plot(title='Quarterly Revenue', grid=True)
```

```python
# Calculating 3-month rolling average revenue
df['3_Month_Rolling_Avg'] = df['Revenue'].rolling(window=3).mean()
print(df.round(2))

```

```python
# Plotting the Revenue and 3-Month Rolling Average
df[['Revenue', '3_Month_Rolling_Avg']].plot(title="Monthly Revenue and 3-Month Rolling Average", grid=True)
```

```python
# Step 1: Import pandas and create a DataFrame with the daily closing price data

import pandas as pd

# Note: This dataset includes daily closing prices for every calendar day in January 2023, including weekends.
# In practice, financial markets are usually closed on weekends and holidays.
# This is a simplified example for demonstration purposes.

data = {
    'Date': [
        '2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
        '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10',
        '2023-01-11', '2023-01-12', '2023-01-13', '2023-01-14', '2023-01-15',
        '2023-01-16', '2023-01-17', '2023-01-18', '2023-01-19', '2023-01-20',
        '2023-01-21', '2023-01-22', '2023-01-23', '2023-01-24', '2023-01-25',
        '2023-01-26', '2023-01-27', '2023-01-28', '2023-01-29', '2023-01-30',
        '2023-01-31'
    ],
    'Closing Price': [
        150, 152, 148, 149, 151, 153, 154, 155, 157, 156,
        158, 160, 162, 161, 159, 158, 157, 156, 155, 153,
        154, 156, 157, 158, 159, 160, 162, 161, 160, 158,
        159
    ]
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
# Step 2: Convert the 'Date' column to datetime format and set it as the index

df['Date'] = pd.to_datetime(df['Date'])  # Convert strings to datetime objects
df.set_index('Date', inplace=True)       # Set the 'Date' column as the index
```

```python
# Step 3: Plot the daily closing price

import matplotlib.pyplot as plt

df['Closing Price'].plot(title="Daily Closing Price - January 2023", grid=True)
```

```python
# Step 4: Resample the data to weekly frequency and calculate average closing price

weekly_closing = df['Closing Price'].resample('W').mean()  # Weekly average
print(weekly_closing)  # Display the resampled data
```

```python
# Step 5: Plot the weekly average closing prices

weekly_closing.plot(title='Weekly Average Closing Price', grid=True)
```

```python
# Step 6: Calculate a 5-day rolling average to smooth short-term variations

df['5_Day_Rolling_Avg'] = df['Closing Price'].rolling(window=5).mean()
print(df.round(2))  # Print the DataFrame with the new column, rounded to 2 decimal places
```

```python
# Step 7: Plot both the original closing price and the 5-day rolling average

df[['Closing Price', '5_Day_Rolling_Avg']].plot(title="Daily Closing Price and 5-Day Rolling Average", grid=True)
```
